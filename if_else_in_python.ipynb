```python
if 20.5: # Non- zero  value print if condition
  print("Correct")
else:
  print("Not")
```

```python
if -20: # Negative value print if condition
  print("Correct")
else:
  print("Not")
```

```python
if 0.0: # float value print else condition
  print("Correct")
else:
  print("Not")
```

```python
if True:
  print("Correct")
else:
  print("Not")
```

```python
if False:
  print("Correct")
else:
  print("Not")
```

```python
# if list is empty then always result is (else)

if []:
  print("Correct")
else:
  print("Not")
```

```python
# Non empty list result is (if)

if [1,2]:
  print("Correct")
else:
  print("Not")
```

```python
# set is empty result is (else)

if {}:
  print("Correct")
else:
  print("Not")
```

```python
# set is Non-empty result is (if)

if {1,2,3}:
  print("Correct")
else:
  print("Not")
```

```python
# Tuple is empty then result is (else)

if ():
  print("Correct")
else:
  print("Not")
```

```python
# Tuple is Non- empty then result is (if)

if (1,23,4):
  print("Correct")
else:
  print("Not")
```
